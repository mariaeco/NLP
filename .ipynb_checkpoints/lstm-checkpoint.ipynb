```python
def split_tokens(row):
    row['tokens'] = word_tokenize(row['text'].lower())
    return row
```

```python
dataset = dataset.map(split_tokens)
```

```python
print(dataset['train'][0]['tokens'])
```

```python
all_tokens = []

for tokens in dataset['train']['tokens']:
  all_tokens.extend(tokens)
```

```python
from collections import Counter
counts = Counter(all_tokens)
```

```python
counts.most_common(50)
```

```python
vocab = [token for token, count in counts.items() if count >= 3]
```

```python
def convert_tokens(tokens):
  result = []
  for token in tokens:
    result.append(token2id.get(token, 0))

  return result
```

```python
print(convert_tokens(dataset['train'][0]['tokens']))
```

```python
def tokens2ids(row):
  row['tokens_id'] = convert_tokens(row['tokens'])
  return row
```

```python
dataset = dataset.map(tokens2ids)
```

```python
print(dataset['train'][0]['tokens_id'])
```

```python
import torch
from torch.utils.data import Dataset

class IMDBDataset(Dataset):
  def __init__(self, dataset):
    self.data = dataset['tokens_id']
    self.labels = dataset['label']

  def __len__(self):
    return len(self.labels)

  def __getitem__(self, idx):
    return torch.tensor(self.data[idx]), torch.tensor(self.labels[idx])
```

```python
train_dataset = IMDBDataset(dataset['train'])
```

```python
len(train_dataset)
```

```python
train_dataset[12499]
```

```python
from torch.utils.data import DataLoader

dataloader = DataLoader(train_dataset, shuffle=True)
```

```python
from torch import nn

EMBED_LEN = 128
HIDDEN_DIM = 32
N_LAYERS = 2


class TextClassifier(nn.Module):
  def __init__(self):
    super().__init__()
    self.embedding_layer = nn.Embedding(num_embeddings=len(vocab), embedding_dim=EMBED_LEN)
    self.rnn = nn.LSTM(input_size=EMBED_LEN, hidden_size=HIDDEN_DIM, num_layers=N_LAYERS)
    self.linear = nn.Linear(HIDDEN_DIM, 1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, input):
    # print('input', input)
    # print('input shape', input.shape)

    embeddings = self.embedding_layer(input)
    # print('embeddings', embeddings)
    # print('embeddings shape', embeddings.shape)

    output, (final_hidden, final_cell) = self.rnn(embeddings)

    # print('output: ', output)
    # print('output shape: ', output.shape)

    # print('final_hidden: ', final_hidden)
    # print('final_hidden shape: ', final_hidden.shape)

    # print('final_cell: ', final_cell)
    # print('final_cell shape: ', final_cell.shape)

    return self.sigmoid(self.linear(final_hidden[-1]))
```

```python
model = TextClassifier()
```

```python
train_dataset[0][0]
```

```python
model(train_dataset[0][0])
```

```python
LR = 1e-3
EPOCHS = 5
loss_fn = nn.BCELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=LR)
```

```python
from tqdm.notebook import tqdm

def train(dataloader):
  size = len(dataloader.dataset)
  train_loss = []

  for batch, (data, label) in enumerate(tqdm(dataloader)):
    data, label = data.to(device), label.to(device)
    # print('data', data)
    # print('data shape', data.shape)

    # print('label', label)
    # print('label shape', label.shape)

    out = model(data[0])
    # print('out', out)
    # print('out shape', out.shape)

    loss = loss_fn(out, label.to(torch.float32))
    # print('loss', loss)
    # print('loss shape', loss.shape)


    train_loss.append(loss.item())

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if batch % 1000 == 0:
      loss, current = loss.item(), batch * len(data)
      print(f' loss: {loss:>7f}  [{current:>5d}/{size:>5d}]')


  # train_loss /= size
  return train_loss
```

```python
def test(dataloader):
  size = len(dataloader.dataset)
  test_loss = []
  correct = 0

  with torch.no_grad():
    for data, label in tqdm(dataloader):
      data, label = data.to(device), label.to(device)
      out = model(data[0])

      loss = loss_fn(out, label.to(torch.float32))
      test_loss.append(loss.item())

      label_out = 1 if out > 0.5 else 0
      if label_out == label[0]:
        correct += 1

  # test_loss /= size
  correct /= size
  print(f' test accuracy: {(100*correct):>0.1f}%, test loss: {test_loss[-1]:>8f} ')

  return test_loss
```

```python
device = torch.device('cuda')
model.to(device)

test_dataset = IMDBDataset(dataset['test'])
dataloader_test = DataLoader(test_dataset, shuffle=True)
```

```python
train_losses = []
test_losses = []

for t in range(EPOCHS):
    print(f'Epoch {t+1} -------------------------------')

    print('Train')
    train_losses.append(train(dataloader))

    print('Test')
    test_losses.append(test(dataloader_test))
```

```python
flat_train_losses = train_losses[0] + train_losses[1] + train_losses[2] + train_losses[3] + train_losses[4]
flat_test_losses = test_losses[0] + test_losses[1] + test_losses[2] + test_losses[3] + test_losses[4]
```
