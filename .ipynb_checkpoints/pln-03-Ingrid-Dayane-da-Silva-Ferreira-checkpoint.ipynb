### Preparando os Dados

```python
import datasets
dataset = datasets.load_dataset("okite97/news-data")
```

```python
dataset
```

O dataset é dividido em dois conjuntos: train e test
Contém também 3 features: uma que contém o Título da Notícia ('Title'), uma que tem um trecho da notícia ('Excerpt') e outra que contém a Categoria da Notícia ('Category').

Vamos criar uma coluna para numerar a categoria.

```python
import datasets
```

As clasificações receberam valores discretos entre 0 e 5 para a função CrosEntropyLoss que será usada mais a diante.

```python
def gera_id_categ(dataset):
    categoria = dataset['Category']
    if categoria == 'business':
        dataset['label'] = 0
    elif categoria == 'entertainment':
        dataset['label'] = 1
    elif categoria == 'health':
        dataset['label'] = 2
    elif categoria == 'politics':
        dataset['label'] = 3
    elif categoria == 'sports':
        dataset['label'] = 4
    elif categoria == 'tech':
        dataset['label'] = 5
    return dataset
```

```python
dataset = dataset.map(gera_id_categ)
```

```python
print(dataset['train']['label'][0]) # teste
```

Após criadas a coluna no Dataset, vamos dividir em Tokens.

### Tokenizando

```python
!pip install torchtext==0.17.0
```

```python
from torchtext.data.utils import get_tokenizer
tokenizer = get_tokenizer('basic_english')
```

Tokenizando os trechos das notícias.

```python
def split_tokens(row):
    row['tokens'] = tokenizer(row['Excerpt']) # cria uma nova coluna chamada tokens e armazena a coluna concat tokenizada
    return row
dataset = dataset.map(split_tokens) # aplicando a função split_tokens para todo o dataset
```

```python
from torchtext.vocab import build_vocab_from_iterator
vocab_custom = build_vocab_from_iterator(dataset['train']['tokens'], min_freq=3, specials=[''])
vocab_custom.set_default_index(vocab_custom['']) # define <unk> como defaut
```

```python
len(vocab_custom) # tamanho do vocabulário
```

```python
def vocab(tokens):
  result = []
  for token in tokens:
    result.append(vocab_custom[token])

  return result
```

```python
def token2id(row):
    row['tokens_id'] = vocab(row['tokens'])
    return row
```

```python
dataset = dataset.map(token2id)
```

### Criando o Dataset e o DataLoader

```python
from torch.utils.data import Dataset, DataLoader
import torch
```

```python
class IMDBDataset(Dataset):
    def __init__(self, dataset): # construtor
        self.data = dataset['tokens_id']
        self.labels = dataset['label']

    def __len__(self): # Qtd de exemplos
        return len(self.data)

    def __getitem__(self, idx): # Acessa um exemplo
      return torch.tensor(self.data[idx]), torch.tensor(self.labels[idx])
```

```python
dataloader = DataLoader(IMDBDataset(dataset['train']), shuffle=True)
```

### Rede Neural

Aqui vamos criar a rede neural usando o Pytorch. Inicialmente usaremos um embedding com retorno de 32 dimensões, a camada intermediária com um retorno de 8 dimensões e ai final a saída possui 1 dimensão.

```python
from torch import nn

EMBED_LEN = 32
HIDDEN_DIM = 8
N_LAYERS = 1

class Classificador(nn.Module):
  def __init__(self):
    super(Classificador, self).__init__()

    self.embedding_layer = nn.Embedding(num_embeddings=len(vocab_custom),embedding_dim=EMBED_LEN)
    self.rnn = nn.RNN(input_size=EMBED_LEN, hidden_size=HIDDEN_DIM, num_layers=N_LAYERS)
    self.linear = nn.Linear(HIDDEN_DIM,6)
    self.CrossEntropyLoss = nn.CrossEntropyLoss()

  def forward(self, input, target):
    embeddings = self.embedding_layer(input)
    output, final_hidden = self.rnn(embeddings)
    return self.linear(final_hidden[0])
```

```python
model = Classificador()
model(torch.tensor(dataset['train']['tokens_id'][0]), torch.tensor(dataset['train']['label'][0]))
```

### Treinamento

O treinamento será executado em CPU.

```python
device = torch.device('cpu')
model.to(device)
```

```python
LR = 1e-5
EPOCHS = 5
optimizer = torch.optim.Adam(model.parameters(), lr=LR)
```

No treinamento e no teste será passado o dataset com a coluna de classificação ('label').


```python
from tqdm.notebook import tqdm

def train(dataloader):
  size = len(dataloader.dataset)
  train_loss = 0

  for batch, (data, label) in enumerate(tqdm(dataloader)):
      data, label = data.to(device), label.to(device) # passando a lista de tokens_ids e seus labels para a CPU
      out = model(data[0],label) # lista de tokens para classificar e aprender
      loss = model.CrossEntropyLoss(out.squeeze(0), label.squeeze().to(torch.long)) # converte para float para a função loss
      train_loss += loss.item() # acumulando o erro

      # Backpropagation e atualização dos pesos
      optimizer.zero_grad()
      loss.backward()
      optimizer.step()

      # A cada 500 iterações exibir o status do treinamento.
      if batch % 500 == 0:
          loss, current = loss.item(), batch * len(data)
          print(f' loss: {loss:>7f}  [{current:>5d}/{size:>5d}]')

  train_loss /= size # média do erro

  return train_loss
```

```python
def test(dataloader):
  size = len(dataloader.dataset)
  test_loss = 0
  correct = 0

  with torch.no_grad(): # desativando o gradiente
    for data, label in tqdm(dataloader):
      data, label = data.to(device), label.to(device)
      out = model(data[0],label)

      loss = model.CrossEntropyLoss(out.squeeze(0), label.squeeze().to(torch.long)) # converte para float para a função loss
      test_loss += loss.item()

      label_out = torch.argmax(out).to(torch.long)

      if label_out == label[0]:
        correct += 1


  test_loss /= size  # Média do erro
  correct /= size  # Acurácia

  print(f' test accuracy: {(100*correct):>0.1f}%, test loss: {test_loss:>8f} ')

  return test_loss
```

```python
dataloader_train = DataLoader(IMDBDataset(dataset['train']), shuffle=True)
dataloader_test = DataLoader(IMDBDataset(dataset['test']), shuffle=True)
```

```python
train_losses = []
test_losses = []
for t in range(EPOCHS):
    print(f'Epoch {t+1} -------------------------------')

    print('Train')
    train_losses.append(train(dataloader_train))

    print('Test')
    test_losses.append(test(dataloader_test))
```

```python
import matplotlib.pyplot as plt
plt.plot(train_losses, label='Train Loss')
plt.plot(test_losses, label='Test Loss')
plt.legend()
plt.show()
```
