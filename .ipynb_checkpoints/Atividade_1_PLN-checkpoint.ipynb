```python
import re
import fitz  # PyMuPDF
from collections import Counter
import random
from collections import defaultdict
from nltk import trigrams
from nltk.tokenize import word_tokenize
import nltk
```

#USANDO LIVROS DE AGATHA CHRISTIE PARA CONTEXTO DE GERAÇÃO DE TEXTO.

```python
def extrair_texto_pdf(caminho_pdf):
    texto = ""
    with fitz.open(caminho_pdf) as doc:
        for pagina in doc:
            texto += pagina.get_text("text") + "\n"
    return texto

# Exemplo de uso
texto = "LIVRO1.pdf"
texto_extraido = extrair_texto_pdf(texto)
texto1 = texto_extraido.lower()

texto = "LIVRO2.pdf"
texto_extraido = extrair_texto_pdf(texto)
texto2 = texto_extraido.lower()

texto = "LIVRO3.pdf"
texto_extraido = extrair_texto_pdf(texto)
texto3 = texto_extraido.lower()


texto = texto1 + "\n" + texto2 + "\n" + texto3
print(texto[500:1000])#MOSTRANDO PARTE DO TEXTO
```

```python
def preprocess_text(texto):
    regex = r'\w+|[,!\.\?;:\/\-\—\"]+'
    texto= re.findall(regex, texto)
    return texto


#SEPARANDO OS TRIGRAMAS E CALCULANDO DUAS PROBABILIDADES
def modelo_trigrama(words):
    #juntando as 3 palavras
    trigrams = [(words[i], words[i+1], words[i+2]) for i in range(len(words) - 2)]
    # Contagem dos trigramas
    trigrama_counts = Counter(trigrams)
    # Contagem dos bigramas
    bigrama_counts = Counter([(w1, w2) for w1, w2, _ in trigrams])
    size = len(set(words))
    #calculando a probabilidade do trigrama
    trigram_prob = {}
    for (w1, w2, w3), count in trigrama_counts.items():
        trigram_prob[(w1, w2, w3)] = (count) / (bigrama_counts.get((w1, w2), 0))
    return trigram_prob

#GERANDO O TEXTO COM BASE NAS PROBABILIDADES
def gerar_texto(model, words, length):
    bigrams = list(set([(w1, w2) for (w1, w2, w3) in model.keys()]))
    #escolha aleatória da palavra
    choosen_word = random.choice(words)
    #Encontrar os bigramas que começam com essa palavra
    bigrams = [bigrama for bigrama in set((w1, w2) for (w1, w2, w3) in model.keys()) if bigrama[0] == choosen_word]
    #escolher aleatoriamente um desses bigramas
    choosen_bigram = random.choice(bigrams)
    newtext = list(choosen_bigram)

    #escolhendo os possíveis trigramas que possuem o bigrama escolhido
    for _ in range(length):
        candidates = [(w1, w2, w3) for (w1, w2, w3) in model if (w1, w2) == choosen_bigram]
        if not candidates:
            break

        #escolhendo a próxima palavra com base na probabilidade
        wordslist = [w3 for (_, _, w3) in candidates]
        probs = [model[t] for t in candidates]
        next_word = random.choices(wordslist, weights=probs)[0]

        newtext.append(next_word)
        choosen_bigram = (choosen_bigram[1], next_word)

    return ' '.join(newtext)

#IMPRIMINDO TEXTO FORMATADO
def formatar_texto(texto):
    frases = texto.split()
    paragrafo = []
    texto_formatado = ""

    for palavra in frases:
        paragrafo.append(palavra)
        if palavra.endswith((".", "!", "?")):
            texto_formatado += " ".join(paragrafo) + "\n\n"
            paragrafo = []

    if paragrafo:
        texto_formatado += " ".join(paragrafo)

    return texto_formatado
```

```python
words = preprocess_text(texto)
trigram_model = modelo_trigrama(words)

text_size =  1000 #quantidade de palavras que desejo
texto_gerado = gerar_texto(trigram_model, words, text_size)
texto_formatado = formatar_texto(texto_gerado)
print(texto_formatado)
```
